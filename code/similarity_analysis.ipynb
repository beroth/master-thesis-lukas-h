```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from config import get_gradient_similarity_file_path
```

```python
def plot_heatmap(data: pd.DataFrame):
    ax = plt.axes()
    sns.heatmap(data, cmap="YlGnBu", ax=ax, annot=True)
    ax.set_xlabel("paraphrased_id")
    
    plt.show()
```

```python
gradient_similarities_subset_5 = pd.read_csv(filepath_or_buffer=get_gradient_similarity_file_path(5), index_col=0)
```
